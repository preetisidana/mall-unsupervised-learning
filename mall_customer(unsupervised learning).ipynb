# K-means clustering

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=5)
label=kmeans.fit_predict(x)

print(label)
```

```python
# filtering
plt.scatter(x.loc[label==0,"Annual Income (k$)"],x.loc[label==0,"Spending Score (1-100)"])
plt.scatter(x.loc[label==1,"Annual Income (k$)"],x.loc[label==1,"Spending Score (1-100)"],color='red')
plt.scatter(x.loc[label==2,"Annual Income (k$)"],x.loc[label==2,"Spending Score (1-100)"],color='blue')
plt.scatter(x.loc[label==3,"Annual Income (k$)"],x.loc[label==3,"Spending Score (1-100)"],color='black')
plt.scatter(x.loc[label==4,"Annual Income (k$)"],x.loc[label==4,"Spending Score (1-100)"],color='pink')
```

# Hierarchical clustering

```python
from scipy.cluster import hierarchy as sch

sch.dendrogram(sch.linkage(x,method="ward"))
plt.xlabel("points")
plt.ylabel("distance")
plt.title('Dendogram')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=5)

labels  = agg.fit_predict(x)
print(labels)
```
